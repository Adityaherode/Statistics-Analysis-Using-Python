# CI

1. Smartphone Screen Time

```python
a=[9.8, 10.3, 10.1, 9.9, 10.4,10.2, 9.7, 10.0, 10.5, 9.8,10.1,
9.9, 10.2, 10.3, 10.0,9.8, 10.4, 10.1, 9.9, 10.2]
```

```python
mn=np.mean(a)
mn
```

```python
std= np.std(a)
std
```

```python
n=20
dof= n-1
dof
```

```python
tscore= t.ppf(0.95,dof)
tscore
```

```python
moe= tscore*(std/np.sqrt(n))
moe
```

```python
ci= mn-moe, mn+moe
ci
```

# Hypothesis

1. AutoDrive Components Ltd. – Engine Assembly Time
Evaluation

```python
#H0
#H1
```

```python
theory_mean = 1.8+0.02*450
mean= 10.8
std = 25
sample = 35
alpha = 0.05
```

```python
#tstates
```

```python
standard_error = std/np.sqrt(sample)
t_stats = (mean - theory_mean) / standard_error 
t_stats
```

```python
critical_value = norm.ppf(1-alpha)
critical_value
```

```python
if t_stats > critical_value:
    decision = 'h1'
else:
    decision = 'h0'
```

```python
decision
```
